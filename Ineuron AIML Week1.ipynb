### 2.Python Program 1.

Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line. 

```python
# Create an Empty List.
op_list = []

# Assign the Lower_range and Upper Range.
lower_range = 2000
upper_range = 3200 + 1  # Upper Range is not included, so to include 3200 added 1.

# Use for loop and get the Numbers into the Output list(Numbers are Divisible by 7 and not divisible by 5).
for i in range(lower_range,upper_range):
    if i % 7 == 0:
        if i % 5 != 0:
            op_list.append(i)

# Print the list.           
print(op_list)
```

```python
# Get the Length of the Output list.
len(op_list)
```

There are 138 numbers between 2000 and 3200 that are divisible by 7 and not divisible by 5.

### 4.Python Program 3.

Write a Python program to find the volume of a sphere with diameter 12 cm.  
 
Formula: V=4/3 * π * r 3 

```python
# import math package.
import math

# Radius = Diameter/2.
diameter = 12
rad = diameter / 2

# Get the Volume of the Sphere.
Vol = ((4/3) * math.pi * (rad**3))

# Print the Volume.
print('Volume of the sphere with diameter 12 cm is {0:.2f} cm\u00b3.'.format(Vol))
```

### 1.Python Program 1.

Write a program which accepts a sequence of comma-separated numbers from console and generate a list. 

```python
# Get the List of Numbers from Console(Numbers are comma Separated).
str1 = input('Provide the List of Numbers -')

# Create an Empty List.
mylist = []
decnum = ' '   # Create a local String variable.

# Use For Loop and get the Numbers.
for i in range(len(str1)):
# Number can be integer or Float, So getting the number in full(with decimal places if available) till we encounter a ','.   
    if decnum == ' ':
        decnum = str1[i]
    elif str1[i] != ',':
        decnum = decnum + str1[i]
        
# Comma Encountered, So get the number into the List.        
    if str1[i] == ',':
        ret = decnum.find('.')
        if ret == -1:
            num = int(decnum)
        else:
            num = float(decnum)
        mylist.append(num)
        decnum = ' '

# Last Number Appending happens, as ',' are not encountered at the end.
# if Commas encountered also, decnum will be initial.
if decnum != ' ':
    ret = decnum.find('.')
    if ret == -1:
        num = int(decnum)
    else:
        num = float(decnum)
    mylist.append(num)
        
# Print the List of Numbers.
print(mylist)
```

Create the below pattern using nested for loop in Python.

```python
for i in range(0, 4):
    for j in range(0, i + 1):
        print("* ", end="")
    print("\r")
for i in range(4, -1 , -1):
    for j in range(0, i + 1):
        print("* ", end="")
    print("\r")       
```
